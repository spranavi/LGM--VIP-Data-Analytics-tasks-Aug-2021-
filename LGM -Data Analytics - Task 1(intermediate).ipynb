#### Exploratory Data Analysis on Dataset - Terrorism

##### 1.Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from tabulate import tabulate
```

##### 2.Reading the data

```python
data = pd.read_csv("globalterrorismdb_0718dist.csv", encoding="ISO-8859-1")
print(data)
```

##### 3.Initial data exploration

```python
data.info()
```

```python
data.columns
```

```python
cols = list(data.columns)
print(cols)
```

```python
nrows =data.shape[0]
for col in cols:
    nulvalues=data[col].isna().sum()
    print(col,data[col].isna().sum())
    
    if nulvalues==nrows:
        data = data.drop([col])
data.shape
```

```python
data.describe()
```

```python
data.corr()
```

##### 4.Data Cleaning

```python
col_names = ["eventid","iyear","imonth","iday","country","country_txt","region","region_txt","provstate","city","latitude","longitude","specificity","vicinity","success","suicide","attacktype1","attacktype1_txt","targtype1","targtype1_txt","target1","natlty1","natlty1_txt","gname","motive","weaptype1","weaptype1_txt","nkill","nwound","property","propextent","propextent_txt","propvalue","ransom","ransomamt","ransompaid"
]
data = data.reindex(columns=col_names)
```

```python
data.head(5)
```

```python
data.fillna(0)
```

##### 5.Exploratory data analysis

##### (i) Attacks per year

```python
columnsTitles=['iyear', 'Count']
df_temp = data.reindex(columns=columnsTitles)
df_temp['Count'] = 1
df_br1 = df_temp.groupby(['iyear'], as_index = True).sum()
plt.figure()
plt.rcParams['figure.figsize'] = [15, 5]
df_br1.plot.bar()
df_br1 = df_br1.sort_values(['Count'], ascending = False)
print(tabulate(df_br1, tablefmt="grid", headers="keys"))
```

###### MAXIMUM NUMBER OF ATTACKS WERE IN THE YEAR 2014

##### (ii)Attacks per Country

```python
columnsTitles=['country_txt', 'Count']
df_temp = data.reindex(columns=columnsTitles)
df_temp['Count'] = 1
df_br1 = df_temp.groupby(['country_txt'], as_index = True).sum()
plt.figure()
plt.rcParams['figure.figsize'] = [15, 5]

df_br1 = df_br1.sort_values(['Count'], ascending = False)
print(tabulate(df_br1, tablefmt="grid", headers="keys"))
df_br1[:10].plot.bar()
```

###### MAXIMUM ATTACKS WERE IN THE COUNTRY IRAQ

##### (iii)Attacks per Region

```python
columnsTitles=['region_txt', 'Count']
df_temp = data.reindex(columns=columnsTitles)
df_temp['Count'] = 1
df_br1 = df_temp.groupby(['region_txt'], as_index = True).sum()
plt.figure()
plt.rcParams['figure.figsize'] = [15, 5]
df_br1.plot.bar()
df_br1 = df_br1.sort_values(['Count'], ascending = False)
print(tabulate(df_br1, tablefmt="grid", headers="keys"))
```

###### MAXIMUM ATTACKS WERE IN MIDDLEEAST AND NORTH AFRICA REGION

##### (iv)Types of Weapons used

```python
columnsTitles=['weaptype1_txt', 'Count']
df_temp = data.reindex(columns=columnsTitles)
df_temp['Count'] = 1
df_br1 = df_temp.groupby(['weaptype1_txt'], as_index = True).sum()
plt.figure()
plt.rcParams['figure.figsize'] = [15, 5]
df_br1.plot.bar()
df_br1 = df_br1.sort_values(['Count'], ascending = False)
print(tabulate(df_br1, tablefmt="grid", headers="keys"))
```

###### EXPLOSIVES WERE THE MOST USED TYPE OF WEAPON IN THESE ATTACKS

##### (v)Group responsible for maximum attacks

```python
columnsTitles=['gname', 'Count']
df_temp = data.reindex(columns=columnsTitles)
df_temp['Count'] = 1
df_br1 = df_temp.groupby(['gname'], as_index = True).sum()
plt.figure()
plt.rcParams['figure.figsize'] = [15, 5]

df_br1 = df_br1.sort_values(['Count'], ascending = False)
print(tabulate(df_br1[:10], tablefmt="grid", headers="keys"))
df_br1[:10].plot.bar()
```

###### KNOWN GROUP RESPONSIBLE FOR MAXIMUM ATTACKS IS TALIBAN

##### (vi)Most affected Target types

```python
columnsTitles=['targtype1_txt', 'Count']
df_temp = data.reindex(columns=columnsTitles)
df_temp['Count'] = 1
df_br1 = df_temp.groupby(['targtype1_txt'], as_index = True).sum()
plt.figure()
plt.rcParams['figure.figsize'] = [15, 5]

df_br1 = df_br1.sort_values(['Count'], ascending = False)
print(tabulate(df_br1[:10], tablefmt="grid", headers="keys"))
df_br1[:10].plot.bar()
```

###### PRIVATE CITIZENS AND PROPERTY WERE MOST AFFECTED TARGETS IN THESE ATTACKS

##### (vii)Weapon type used in each year

```python
pd.crosstab(data["iyear"], data["weaptype1_txt"]).plot(kind='area',stacked=False,figsize=(20,10))
plt.title('Weapon type used In Each Year',fontsize=25)
plt.ylabel('Number of Attacks',fontsize=20)
plt.xlabel("Year",fontsize=20)
plt.show()
```
