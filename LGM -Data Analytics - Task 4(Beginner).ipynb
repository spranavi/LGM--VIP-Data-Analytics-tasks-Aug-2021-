```python
import matplotlib.pyplot as plt
import cv2
```

##### 2.Reading the original image

```python
img = plt.imread("image.jfif")
plt.figure(figsize=(8,8))
plt.imshow(img)
plt.axis("off")
plt.title("Actual Image")
plt.show()
```

##### 3.RGB to grayscale conversion

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(8,8))
plt.imshow(gray_img,cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.show()
```

##### 4.Negative Conversion

```python
invert_img = cv2.bitwise_not(gray_img)
plt.figure(figsize=(8,8))
plt.imshow(invert_img,cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.show()
```

##### 5.Pencil Sketch Creation

```python
smoothing_img = cv2.GaussianBlur(invert_img, (25, 25),sigmaX=0, sigmaY=0)
```

```python
final = cv2.divide(gray_img, 255 - smoothing_img, scale=256)
plt.figure(figsize=(8,8))
plt.imshow(final,cmap="gray")
plt.axis("off")
plt.title("Final Sketch Image")
plt.show()
```
